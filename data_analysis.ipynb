```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
loss_1e_3_16 = []
with open('loss/loss_1e_3_16.txt','r',encoding = 'utf-8') as f:
    lines = f.readlines()
    for line in lines:
        if 'Loss' in line:
            loss = line.split('||')[1].strip()
            loss_1e_3_16.append(float(loss[6:]))
```

```python
# 定义颜色
color1 = '#0085c3'
color2 = '#7ab800'
color3 = '#dc5034'
color4 = '#FFA500'

# 设置图像大小
fig = plt.figure(figsize=(20, 12))

# 绘制折线图
step = 1
maxiteration = 6000
x = range(0, maxiteration, step)
plt.subplot(221)
plt.plot(x, loss_1e_3_32[:maxiteration:step], color=color1, label='loss_1e_3_32')
plt.title('loss when batch_size = 32', fontsize=15)
plt.grid(ls=':', color='gray', alpha=0.6)
plt.legend(loc='upper left', fontsize=12)
plt.xlabel("iteration")
plt.ylabel("loss")

plt.subplot(222)
plt.plot(x, loss_1e_3_16[:maxiteration:step], color=color2, label='loss_1e_3_16')
plt.title('loss when batch_size = 16', fontsize=15)
plt.grid(ls=':', color='gray', alpha=0.6)
plt.legend(loc='upper left', fontsize=12)
plt.xlabel("iteration")
plt.ylabel("loss")

plt.subplot(223)
plt.plot(x, loss_5e_4_4[:maxiteration:step], color=color3, label='loss_5e_4_4')
plt.title('loss when batch_size = 4', fontsize=15)
plt.grid(ls=':', color='gray', alpha=0.6)
plt.xlabel("iteration")
plt.ylabel("loss")
plt.legend(loc='upper left', fontsize=12)

plt.subplot(224)
plt.plot(x, loss_5e_4_2[:maxiteration:step], color=color4, label='loss_5e_4_2')
plt.title('loss when batch_size = 2', fontsize=15)
plt.grid(ls=':', color='gray', alpha=0.6)
plt.xlabel("iteration")
plt.ylabel("loss")
plt.legend(loc='upper left', fontsize=12)
```

```python
mAP_1e_3_32 = []
file_list = []
file_path = os.listdir('weights/1e_3_32')
for file in file_path:
    if file.startswith('ssd') or file.startswith('E'):
        continue
    file_list.append(int(file.split('.')[0]))
    with open(os.path.join('weights/1e_3_32', file),'r',encoding = 'utf-8') as f:
        lines = f.readlines()
        for line in lines:
            if line.startswith("Mean AP ="):
                mAP = float(line.split('=')[-1].strip())
                mAP_1e_3_32.append(mAP)
print(mAP_1e_3_32)
print(file_list)
data1 = pd.DataFrame([])
data1['mAP'] = mAP_1e_3_32
data1['iteration'] = file_list
data1 = data1.sort_values(['iteration'],ascending=True)
data1
```

```python
# 定义颜色
color1 = '#0085c3'
color2 = '#7ab800'
color3 = '#dc5034'
color4 = '#FFA500'

# 设置图像大小
fig = plt.figure(figsize=(15, 8))

# 绘制折线图
plt.subplot(111)
plt.plot(data1['iteration'], data1['mAP'], color=color1, marker='o', label='mAP_1e_3_32')
plt.title('mAP when batch_size = 32', fontsize=15)
plt.grid(ls=':', color='gray', alpha=0.6)
plt.legend(loc='upper left', fontsize=12)
plt.vlines(85000, 0, 0.8, linestyles='-.', colors=color3)
plt.text(75000, 0.6, '85000 0.7776', color=color3, fontsize=10)
plt.xticks([1000, 10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000, 5000, 50000, 55000, 60000, 65000, 70000, 75000, 80000, 87000])
plt.xlabel("iteration")
plt.ylabel("mAP")
```
